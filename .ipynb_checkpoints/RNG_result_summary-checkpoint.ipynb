- A total of $\textbf{10 billion}$ random numbers were generated from five different distributions.
- $\textbf{g++ version 9.4.0}$ from the GNU Compiler Collection was used across all platforms.
- The random numbers were never written to memory, therefore only $\textbf{CPU performances}$ were measured.
